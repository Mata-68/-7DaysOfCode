# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df_data = pd.read_csv('/content/inmigrantes_canada.csv')
df_data.head()
```

```python
df_data.shape
```

```python
df_data.info()
```

```python
#Para configurar el índice, en este caso el país

df_data.set_index('Pais', inplace=True)
df_data
```

```python
# Crear un nuevo dataframe considerando los inmigrantes colombianos

anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df_data.loc['Colombia', anos]
colombia
```

```python
# Creamos una lista para cada año

col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
# Creamos un nuevo DataFrame con los resultados

df_datos_col = pd.DataFrame(col_dict)
df_datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) # PAra asignar tamaño
plt.plot(df_datos_col['Año'], df_datos_col['Inmigrantes']) # Para asignar el nombre de los ejes
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # Para asignar "escala" al eje x
plt.xticks(rotation=90) # para rotar los nombres del eje x1
plt.yticks([1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000]) # Para asignar "escala" al eje y
plt.xlabel('Años') # Para ver el nombre del eje x
plt.ylabel('Inmigrantes') # Para ver el nombre del eje y
plt.title('Inmigrantes de Colombia a Canadá') # Para ver el título del gráfico
plt.show() # Para ver el gráfico.  Este debe ser el último comando
```

```python
# Para poder guardar esta información

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(df_datos_col['Año'], df_datos_col['Inmigrantes']) # Hasta aquí sirve para guardar y desde aquí generar los cambios al gráfico
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.tick_params(axis='x', rotation=90)
ax.yaxis.set_major_locator(plt.MultipleLocator(500))
ax.set_xlabel('Años')
ax.set_ylabel('Inmigrantes')
ax.set_title('Inmigrantes de Colombia a Canadá')
plt.show()
```
